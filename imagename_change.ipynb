```python
import os
import tqdm
from natsort import natsorted

def rename_images(source_folder, dest_folder):
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)

    # 対象のフォルダ内の画像ファイルを取得
    image_files = natsorted([f for f in os.listdir(source_folder) if os.path.isfile(os.path.join(source_folder, f)) and f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif'))])

    for idx, image in enumerate(tqdm.notebook.tqdm(image_files)):
        # 新しい名前を作成する場合：例えば、"image_1.png", "image_2.png"というように命名
        
        new_name = f"{idx + 1:06d}.{image.split('.')[-1]}"
        #new_name = f"image_{idx + 1}.{image.split('.')[-1]}"
        source_path = os.path.join(source_folder, image)
        dest_path = os.path.join(dest_folder, new_name)

        # ファイルの移動と名前の変更
        os.rename(source_path, dest_path)
        print(f"{image} を {new_name} に変更しました。")
```

```python

# ソースフォルダと保存先フォルダを指定して実行
source_directory = '../data/img1'  # 画像ファイルが含まれているフォルダのパスを指定
destination_directory = '../data/11/img1'  # 新しい名前で保存するフォルダのパスを指定

rename_images(source_directory, destination_directory)
```
