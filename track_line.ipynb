```python
import cv2
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
def visualize_trajectory(coordinates):
    # 座標からx軸とy軸の値を取得
    x_values = [coord[0] for coord in coordinates]
    y_values = [coord[1] for coord in coordinates]

    # 新しい図を作成
    plt.figure(figsize=(8, 6))

    # x軸とy軸にプロット
    #plt.xlim(0,1920)
    #plt.ylim(0,1080)
    plt.plot(x_values, y_values, marker='o', linestyle='-')

    # グラフにタイトルと軸ラベルを追加
    plt.title('Tracking Trajectory')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')

    # グラフを表示
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
data = '../output/line_sample.txt'
df = pd.read_csv(data,sep = ',',header=None, names=['frame','id', 'center_x', 'center_y', 'width', 'height','x','y','z'])
print(df)
# X_values列とY_values列から座標を取得しリスト化
coordinates_list = list(zip(df['center_x'], df['center_y']))
    # 座標のリストを可視化
visualize_trajectory(coordinates_list)
```
