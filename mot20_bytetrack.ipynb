## 2. MOT20での実験

### gtが必要なので、trainで実験

```python
from natsort import natsorted

source_dir = '../../mot_datasets/MOT20'

data_list = natsorted(os.listdir(source_dir))
print('movie num = ',len(data_list))
print(data_list)
```

## 3. Detection & Tracking

```python
# Load an official or custom model
from ultralytics import YOLO
model = YOLO('yolov8m.pt') 

#データのインデックスの番号を指定
index = 4

#path指定
path = data_list[index]

print('tracking source = ',path)
print('tracking index  = ',index + 1,'/',len(data_list))

# exp name
exp_dir = '../tracking_results_1114_yolov8_m'
# os.makedirs(exp_dir,exist_ok = True)

exp_name = 'tracking_' + path 
source_path = os.path.join(source_dir,path+'/img1')
# Perform tracking with the model
# visualize --> 特徴量保存
results = model.track(source=source_path, tracker="bytetrack.yaml",save_txt= True, save_conf = True, save = True, 
                      visualize = False,line_width = 2, save_crop = False ,name = os.path.join(exp_dir,exp_name), exist_ok = True)  # Tracking with ByteTrack tracker
```
