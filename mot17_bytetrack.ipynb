## 2. MOT17での実験

```python
from ultralytics import YOLO
from natsort import natsorted

# Load an official or custom model
model = YOLO('yolov8x.pt')  
# Load an official Detect model

source_dir = '../../mot_datasets/MOT17_sub'

data_list = natsorted(os.listdir(source_dir))
print('movie num = ',len(data_list))
print(data_list)
```

## 3. Detection & Tracking

```python
#データのインデックスの番号を指定
index = 9

path = data_list[index]
print('tracking source = ',path)
print('tracking index  = ',index+1,'/',len(data_list))

exp_dir = '../tracking_results_1108_yolov8_x'
# os.makedirs(exp_dir,exist_ok = True)

exp_name = 'tracking_' + path 
source_path = os.path.join(source_dir, path + '/img1')

# Perform tracking with the model
# Tracking with ByteTrack tracker
# visualize --> 特徴量保存

results = model.track(source=source_path, tracker="bytetrack.yaml", save = True, save_txt= True, save_conf = True, 
                      visualize = False, line_width = 2, save_crop = True, name = os.path.join(exp_dir,exp_name),exist_ok = True)  

```
