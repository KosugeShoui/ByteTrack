## dancetrackでの実験

```python
from ultralytics import YOLO
from natsort import natsorted
# Load an official or custom model
model = YOLO('yolov8n.pt')  
# Load an official Detect model

source_dir = '../../mot_datasets/dancetrack/val'

data_list = natsorted(os.listdir(source_dir))
print('movie num = ',len(data_list))
print(data_list)
```

```python
#データのインデックスの番号を指定
index = 9

path = data_list[index]
exp_name = 'tracking_' + path 
source_path = os.path.join(source_dir,path+'/img1')
# Perform tracking with the model
results = model.track(source=source_path, tracker="bytetrack.yaml",save = True,save_txt= True, save_conf = True, line_thickness = 2, save_crop = True,name = exp_name)  # Tracking with ByteTrack tracker
```
